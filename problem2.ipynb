# Student Marks Analysis

You have the marks of 10 students in 3 subjects: Math, Science, and English.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

marks = np.array([
    [78, 85, 90],
    [65, 70, 75],
    [88, 92, 95],
    [55, 60, 58],
    [72, 75, 80],
    [90, 95, 98],
    [68, 72, 70],
    [82, 85, 88],
    [77, 80, 83],
    [60, 65, 68]
])
```

```python
df = pd.DataFrame(marks,columns=["science","maths","english"])
students = [f"students_{i}"for i in range(1,11)]
df["students"] = students
df
print("display first five rows : ", df.head(5))
print("\n\n")
print("display last three rows : ",df.tail(3))
print("\n\n")
# calculate average marks in each subjects
avg_marks = df[["science", "maths", "english"]].mean()
print("average marks in each subjects : ", avg_marks)
print("\n\n")
# add new columns total 
df["Total"] = df["science"] + df["maths"] + df["english"]
# add new columns average 
df["average"] =  df[["science", "maths", "english"]].mean(axis=1)

# find students with highest total marks
student_high_marks = df.loc[df["Total"].idxmax()]
print("students with highest marks :\n ",student_high_marks)
print("\n\n")
# find students with lowest total marks 
student_lowest_marks = df.loc[df["Total"].idxmin()]
print("student with lowest marks : \n",student_lowest_marks)
# summary statistics
df.describe()
df
```

```python
# find overall average marks of all students 
average_marks = np.mean(marks)
print("overall average maks : \n", average_marks)
# find highest marks in the data sets
high_marks = np.max(marks)
print("high marks in the data sets  : \n ", high_marks)
# find lowest marks in the data sets 
lowest_marks = np.min(marks)
print("lowest marks in the data sets : \n", lowest_marks)
# count how many marks are greater than 80
marks_greater_80 = np.sum(marks> 80)
print("marks are greater than 80 : ", marks_greater_80)
# find the standard deviation of all marks
std_all_marks = np.std(marks)
print("standard deviation of all marks : \n", std_all_marks.round(4))
```

```python
subjects = ["science", "maths","english"]
average_marks = df[["science","maths","english"]].mean()
plt.bar(subjects,average_marks,width=0.5,label="std marks",color= "orange")
plt.title("average_marks per subjects")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
# plot the total marks of all students
plt.plot(df["Total"],color="green",marker = "o",linestyle=":")
plt.title("students total marks")
plt.xlabel("stuents numbers")
plt.ylabel("total marks")
plt.grid()
```

# Monthly Sales Analysis

A small shop recorded its sales (in dollars) for 12 months.

```python
sales = np.array([1200, 1350, 1400, 1250, 1500, 1600,
                  1700, 1650, 1800, 1900, 2000, 2100])

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
# Find the total sales for the year.
total_sales = np.sum(sales)
print("total sales for the year : ", total_sales)
# Find the average monthly sales
avg_month_sales = np.mean(sales)
print("monthly sales average : ",avg_month_sales)
# Find the highest sales value.
highest_sale = np.max(sales)
print("highest sales in the months : ",highest_sale)
# Find the lowest sales value.
lowest_sales = np.min(sales)
print("lowest sales in the month : ", lowest_sales)
# Calculate the difference between the highest and lowest sales.
diffrence = highest_sale - lowest_sales
print("diffrence b/w highest sale and lowest sale : ", diffrence)
# Count how many months had sales greater than 1600.
cnt_mont_grt_1600 = np.sum([sales > 1600])
print("counting months have greater sales than 1600 : ", cnt_mont_grt_1600)
# Find the sales growth from one month to the next.
sales_growth = sales[1:] - sales[:-1]
print("sales growth from one month to next  month : ", sales_growth)
```

# Part 2: Pandas Tasks
Create a DataFrame with:
Month
Sales

```python
df = pd.DataFrame({"months":months,"sales":sales})
print(df)
# Profit = Sales * 0.20
df["profit"] = sales * 0.2
# category
def get_category(sale):
        if  sale >= 1800:
            category = "High"
        elif sale >= 1400:
            category = "Medium"
        else:
            category = "Low"
        return category
df["category"] = df["sales"].apply(get_category)
# display first five rows
head = df.head(5)
print("display first five rows : ", head)
# display last three rows
tail = df.tail(3)
print("display last three rows : ", tail)
# display summary statistics
summary_stat = df.describe()
print("display summary statistics : ", summary_stat)
# month with highest sale
high_month_sale =df.loc[ df["sales"].idxmax()]
print("month with highest sale : ",high_month_sale)
# month with lowear sales
lowest_month_sale = df.loc[df["sales"].idxmin()]
print("month with lowest sale : ",lowest_month_sale)
# Total profit for the year
total_profit = np.sum(df["profit"])
print("total profit of the year : ", total_profit)
```
