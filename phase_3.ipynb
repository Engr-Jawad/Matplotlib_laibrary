# pie chart
<p> a pie chart show us how total divided into diffrent parts and show the percentage of each part in total.

```python
import matplotlib.pyplot as plt
```

# autopct = Automatically display percentages on pie chart slices.
%1.0f%% → No decimal places (e.g., 25%)
%1.1f%% → One decimal place (e.g., 25.0%)
%1.2f%% → Two decimal places (e.g., 25.00%)

```python
regions = ["north","south","east","west"]
revenue =[100,200,150,190]
plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=["steelblue","lightgreen","gray","blue"])
plt.title("Revenue contribution by region")
plt.show()
```
