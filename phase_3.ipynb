# pie chart
<p> a pie chart show us how total divided into diffrent parts and show the percentage of each part in total.

```python
import matplotlib.pyplot as plt
```

# autopct = Automatically display percentages on pie chart slices.
%1.0f%% → No decimal places (e.g., 25%)
%1.1f%% → One decimal place (e.g., 25.0%)
%1.2f%% → Two decimal places (e.g., 25.00%)

```python
regions = ["north","south","east","west"]
revenue =[100,200,150,190]
plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=["steelblue","lightgreen","gray","blue"])
plt.title("Revenue contribution by region")
plt.show()
```

# histogram
<P> it is a graph which describe the data but the data is continous and bars touched each other becasue the data is continous.

```python
score = [45,54,65,56,67,76,87,78,98,89,90,80,70,87,65,45,67,65,65,74,85,96,58,69,88,77,67,87,66,99]
plt.hist(score,bins=7,color="gold",edgecolor="black")
plt.title("Distribution of student scores")
plt.xlabel("scores Range")
plt.ylabel("number of students")
plt.show()
# height of bar is equall to number of student in that range
```

```python
# Student Club Membership Distribution
clubs =["programming","sports","musci","arts","robotics"]
members =[80,55,35,20,40]
plt.pie(members,labels= clubs,autopct="%1.2f%%",startangle=90)
plt.title("Student Club Membership Distribution")
plt.axis("equal") # it make the pie chart a perfect circle not the oval 
plt.show()
```

```python
# A bookstore sold the following books in one month:
Category = ["Fiction","Science","History","Technology",	"Comics","Biography"]
book_sold = [250,180,120,220,150,80]
explode=[0,0,0,0.1,0,0]
plt.pie(book_sold,labels = Category,autopct="%1f%%",colors=["red","green","yellow","steelblue","purple","coral"],startangle=140,explode= explode,shadow=True)
plt.title("Monthly Book Sales by Category")
plt.axis("equal")
plt.show()

```
