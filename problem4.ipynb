```python
# How do you plot a line graph?
import matplotlib.pyplot as plt
import pandas as pd
x=[1,2,3,4,5,6]
y=[4,3,2,8,9,7]
plt.plot(x,y)
plt.title("line graph")
plt.show()
```

```python
# How do you plot two lines on the same graph?
x=[1,2,3,4,5]
y1=[3,4,5,6,9]
y2=[3,4,3,6,7]
plt.plot(x,y1,label="line one",color="r",ls=":",marker="o",lw=2.5)
plt.plot(x,y2,label="line two")
plt.legend()
plt.show()
```

```python
# How to plot a time series with dates on X-axis?
dates = pd.date_range("2026-01-03",periods=7)
sales=[7,5,3,8,6,9,4]
plt.Figure(figsize=(10,34))
plt.plot(dates,sales,marker="^",color="b",ls="--")
plt.xlabel("date")
plt.ylabel("sale")
plt.show()
```

```python
marks=[85,72,91,78,88]
students =["ali","khan","zain","shah","ahmad"]
plt.Figure(figsize=(8,6))
plt.bar(students,marks,color="steelblue",edgecolor="black",label="student marks data")
plt.title("student marks ")
plt.xlabel("stuents names")
plt.ylabel("marks")
plt.legend(loc="upper left")
plt.show()
```

```python
# laibrary books
category =["science","history","maths","english","computer science"]
books=[120,80,150,95,170]
plt.Figure(figsize=(10,8))
plt.bar(category,books,color="orange",edgecolor="black")
plt.title("Books by Category")
plt.xlabel("category of books")
plt.ylabel("number of books")
plt.xticks(category,rotation=45,ha="right")
bar = plt.bar(category,books)
plt.bar_label(bar)
plt.grid(axis="y")
plt.show()
```

```python
# Favorite Programming Languages
language = ["Python","Java","C++","JavaScript","Go"]	
students=[220,110,80,60,30]
plt.figure(figsize=(8,6))
plt.bar(language,students,label="student favorite programming language",color= ["gold", "gray","gray","gray","gray"])
plt.title("Favorite programming language")
plt.xlabel("programming language")
plt.ylabel("number of students")
plt.grid(axis='y',ls=":")
plt.xticks(rotation=20)
plt.ylim(0,250)
plt.show()
```

```python
# A supermarket sold the following products in one week.
products=["rice","sugar","milk","bread","eggs","tea"]
sales =[350,280,420,190,510,240]
plt.Figure(figsize=(10,5))
bar = plt.bar(products,sales,label="product sales 2026",color=["red","green","blue","yellow","gray","coral"])
plt.bar_label(bar)
plt.title("Products Sold in One Week")
plt.xlabel("products")
plt.ylabel("sales")
plt.legend(loc="upper left")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
