# Subplots

```python
import matplotlib.pyplot as plt
```

```python


# Correct syntax (standard usage)
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.subplot(1, 2, 1)  # One row, two columns, first subplot
plt.plot(x, y,)
plt.title("Line Chart")

plt.subplot(1, 2, 2)  # Second subplot in the same figure
plt.bar(x, y, color='purple')  # Example: add a bar chart
plt.title("Bar Chart")

plt.tight_layout()  # Adjust spacing so titles don't overlap
plt.show()
```

# professional way 

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
fig,ax = plt.subplots(1,2,figsize=(10,7))
ax[0].plot(x,y,color="green",marker="o",label="data 1 ")
ax[0].set_title("line chart")
ax[1].bar(x,y,color="blue")
ax[1].set_title("bar chart")
plt.suptitle("comparision of line and bar chart")
plt.tight_layout()
plt.show()
```

# Save fig function
use to the pic in the diffrent format like jpeg,png,pdf

```python
# syntax plt.savefig("filename.extension",dpi= value ,bbox_inches ="tight")
months = [1,2,3,4,5,6]
sales = [1222,1332,1423,1111,5432,3241]
plt.plot(months, sales, color = "blue", linestyle = ":",linewidth=1.8,marker = "^",label = "2026 shop sales data")
plt.xlabel("months")
plt.ylabel("sales per month")
plt.title("monthly sales report of a shop")
plt.legend(loc = "upper right")
plt.grid(color = "green",linestyle =":",linewidth = 0.8)
plt.savefig("plot.png",dpi=300)
plt.show()
```
