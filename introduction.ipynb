#### pyplot is a submodule of Matplotlib that provides a collection of functions for creating and customizing plots, charts, and graphs. It offers a simple interface similar to MATLAB for data visualization.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y=[20,10,5,34,23]
plt.plot(x,y,)
plt.show()
```

# key term you should know about the pyplot
<br>Data point -----> a set of values in which one reprsent x axis and another represent y axiss
<br>X axis represent horizantal direction and Y axis represent a vertical direction
<br>Figure -----> whole window
<br>Axes----> cover the area under the graph
<br>plot ---> mean the shape of the graph
<br>marker ---> O for circle and S for square and D for diamond
<br>Linestyle ----> to draw line in the style (-,-.,:)
<br>Color -----> to give the color to charts and graphs (green,blue,black,orange)
<br>

```python
x = ["monday","thuesday","wednesday","thurasday","friday"]
y = [10,23,43,14,33]
plt.plot(x,y,color="green",marker = "o",ms = 14,linestyle = ":",lw=3)
plt.title("shops sales this week ")
plt.xlabel("days of the week")
plt.ylabel("sales per day")
# save figure in the png formate
plt.savefig("my_shop.png")
plt.show()
```

```python
x = np.random.randint(4,9,(3,2))
y = np.random.randint(5,90,(3,2))
plt.plot(x,y,lw =4)
plt.show()
```
