# Weather Analysis Dashboard

```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

temperature = [30, 32, 35, 33, 31, 29, 28]

humidity = [60, 58, 55, 62, 65, 70, 68]

rainfall = [0, 5, 12, 3, 0, 20, 10]

wind_speed = [10, 12, 8, 15, 11, 14, 9]
```

```python
fig,ax = plt.subplots(2,2,figsize=(14,10))
ax[0,0].plot(days,temperature,color="red",marker="o",lw=2.5)
ax[0,0].set_title("Temprature during the week")
ax[0,0].set_xlabel("days")
ax[0,0].set_ylabel("temperature")
ax[0,0].grid(ls=":")

bar = ax[0,1].bar(days,humidity,color="skyblue",edgecolor="black")
ax[0,1].set_title("Humidity Level")
ax[0,1].set_xlabel("days")
ax[0,1].set_ylabel("humidity")
ax[0,1].bar_label(bar)

ax[1,0].scatter(days,rainfall,color = "purple",marker="d")
ax[1,0].set_title("Daily rainfall")
ax[1,0].set_xlabel("days")
ax[1,0].set_ylabel("rainfall")
ax[1,0].grid(ls=":")

ax[1,1].plot(days,temperature,color="red",label="temperature")
ax[1,1].plot(days,wind_speed,color="green",label="wind_speed",ls=":")
ax[1,1].grid(ls=":")
ax[1,1].legend()
ax[1,1].set_title("Temperature vs Wind speed")
ax[1,1].set_xlabel("days")
ax[1,1].set_ylabel("temp and wind speed")

plt.suptitle("Weekly Weather Dashboard")
plt.tight_layout()
plt.savefig("weather_dashboard.png",dpi=300,bbox_inches="tight")
plt.show()


```
