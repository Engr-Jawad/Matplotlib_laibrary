# Student Performance Dashboard

```python
import matplotlib.pyplot as plt
```

```python
students = ["Ali", "Sara", "Ahmed", "Ayesha", "Usman", "Zain"]

math = [78, 85, 92, 88, 76, 95]
science = [82, 79, 94, 90, 81, 97]
english = [75, 88, 91, 84, 80, 93]
attendance = [90, 95, 98, 92, 89, 99]
```

```python
# subplot one 
fig,ax = plt.subplots(2,2,figsize = (10,7))
ax[0,0].bar(students,math,color = "skyblue")
ax[0,0].set_title("math score")
ax[0,0].set_xlabel("students")
ax[0,0].set_label("score")


# subplot two 
ax[0,1].plot(students,science,color = "green",marker="o")
ax[0,1].set_title("science score")
ax[0,1].set_xlabel("students")
ax[0,1].set_ylabel("score")

# subplot three
ax[1,0].scatter(english,attendance,color="red")
ax[1,0].set_title("english score vs attendence")
ax[1,0].set_xlabel("english score")
ax[1,0].set_ylabel("attendance")

#  subplot four
explode=[0,0,0,0,0,0.3]
ax[1,1].pie(math,labels = students,autopct="%1.1f%%",startangle=90,explode=explode)
ax[1,1].set_title("math score distribution")


# dashboard title
plt.suptitle("students Performance  Dashboard")
plt.tight_layout()

plt.show()
```

# Sales Analysis Dashboard

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

sales = [12000, 15000, 18000, 17000, 21000, 25000]

profit = [2500, 3200, 4100, 3900, 5200, 6100]

expenses = [9500, 11800, 13900, 13100, 15800, 18900]

products = ["Laptop", "Phone", "Tablet", "Monitor", "Keyboard"]
quantity = [35, 60, 25, 18, 50]
```

```python
fig,ax = plt.subplots(2,3,figsize=(9,7))
# subplot one
ax[0,0].plot(months,sales,color="green",marker="o")
ax[0,0].set_title("Monthly Sales")
ax[0,0].set_xlabel("months")
ax[0,0].set_ylabel("sales")
ax[0,0].grid()

# subplot two 
ax[0,1].bar(months,profit,color="green",edgecolor= "black")
ax[0,1].set_title("Monthly Profit")
ax[0,1].set_xlabel("months")
ax[0,1].set_ylabel("profit")


# subplot three
ax[0,2].scatter(sales,profit,color="purple")
ax[0,2].set_title("Sales Vs Profit")
ax[0,2].set_xlabel("sales")
ax[0,2].set_ylabel("profit")

#subplot four
ax[1,0].hist(expenses,bins=5,color="green",edgecolor="black")
ax[1,0].set_title("Expenses Distribution")
ax[1,0].set_xlabel("Expenses")
ax[1,0].set_ylabel("Frequency")

#subplot five
ax[1,1].pie(quantity,labels=products,startangle=90,autopct="%1.1f%%")
ax[1,1].set_title("Product Sales Share")

#subpolt six
ax[1,2].barh(products,quantity,color="orange")
ax[1,2].set_title("Units Sold")
ax[1,2].set_xlabel("quantity")
ax[1,2].set_ylabel("products")

plt.suptitle("Sales analysis Dashboard")
plt.tight_layout()
plt.show()
```
