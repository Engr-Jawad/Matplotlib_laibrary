# begginer level problems

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.title("simple line graph")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
y1 = [1,3,5,7,9]
y2 = [2,4,6,8,10]
plt.plot(y1,color = "blue",label = "l1")
plt.plot(y2,color = "green",label = "l2")
plt.legend()
plt.show()
```

```python
Students = ["Ali", "Sara", "Ahmed", "Zara"]
Marks = [75, 85, 60, 90]
plt.bar(Students,Marks)
plt.title("Students Performance")
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1200, 1500, 1800, 1700, 2100, 2500]
plt.plot(months,sales,marker = "o")
plt.title("mothly sales data")
plt.xlabel("months")
plt.ylabel("sales")
plt.grid()
plt.show()
```

```python
courses = ["Python", "SQL", "Excel", "Power BI", "Tableau"]
students = [120, 90, 70, 60, 50]
plt.bar(courses,students,width=0.8,color="green")
plt.xlabel("languages")
plt.ylabel("no of studnents")
```

```python
hours = [1, 2, 3, 4, 5, 6, 7, 8]
scores = [40, 45, 50, 60, 65, 70, 80, 90]
plt.scatter(hours,scores,marker=">")
plt.title("study hours and exam scores")
plt.xlabel("study hours")
plt.ylabel("scores")
```
