```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.title("simple line graph")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
y1 = [1,3,5,7,9]
y2 = [2,4,6,8,10]
plt.plot(y1,color = "blue",label = "l1")
plt.plot(y2,color = "green",label = "l2")
plt.legend()
plt.show()
```
