# data viusalization with matplotlib

```python
import matplotlib.pyplot as plt
```

```python
months = [1,2,3,4,5,6]
sales = [1222,1332,1423,1111,5432,3241]
plt.plot(months, sales, color = "blue", linestyle = ":",linewidth=1.8,marker = "^",label = "2026 shop sales data")
plt.xlabel("months")
plt.ylabel("sales per month")
plt.title("monthly sales report of a shop")
plt.legend(loc = "upper right")
plt.grid(color = "green",linestyle =":",linewidth = 0.8)
plt.show()
```

```python
months = [1,2,3,4,5]
sales =[1000,1200,1800,1500,1100]
plt.plot(months,sales,color = "steelblue",linewidth = 1.3,marker = "d")
plt.xlabel("months")
plt.ylabel("sales per months")
plt.title("2026 sales data")
plt.grid(color="gray",linestyle = ":")
plt.xlim(1,6)
plt.ylim(0,2000)
plt.xticks([1,2,3,4,5],["M1","M2","M3","M4","M5"])
```

# Plotting bar charts , pie charts and histograms 

```python
# bar chart is use for comparsion of two values
product  = ["p1","p2","p3","p4","p5"]
sales = [1200,600,1300,1800,800]
plt.bar(product,sales,color = "yellow",label = "2026 sales data",edgecolor="black") # barh mean horizantal bar 
plt.title("product sales data")
plt.xlabel("products ")
plt.ylabel("sales")
plt.legend()
plt.show()
```
